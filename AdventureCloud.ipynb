```python
from numpy.matrixlib.defmatrix import matrix
from datetime import date
import html
import time
import IPython
import random
import uuid
import json
import base64
import pdb
import time
from PIL import Image
from io import BytesIO
from firebase import firebase
from google.colab import output
from IPython.display import display, HTML
import uuid
import threading
from IPython.display import display, HTML, Javascript
import plotly.graph_objects as go
from IPython.display import display

FIREBASE_CONNECTION = firebase.FirebaseApplication('https://newproj-b2090-default-rtdb.europe-west1.firebasedatabase.app/', None)


class InvokeButtonWithPlace(object):
    def __init__(self, title, callback, source, wSize, hSize, tPos, lPos):
        self._title = title
        self._callback = callback
        self._source = source
        self.wSize = wSize
        self.hSize = hSize
        self.tPos = tPos
        self.lPos = lPos

    def _repr_html_(self):
        callback_id = 'button-' + str(uuid.uuid4())
        output.register_callback(callback_id, self._callback)

        box_shadow = ''
        if self._title == 'rocks':
            box_shadow = 'border-width: 0px; cursor: pointer; background-color: transparent; border-radius: 50px;box-shadow: 0 0 60px 30px rgba(255, 255, 255, 0.4), inset 0 0 6vw 3vw rgba(255, 255, 255, 0.4);'

        template = """
            <style>
                #{callback_id} {{
                    position: absolute;
                    top: {tPos};
                    left: {lPos};
                    width: {wSize};
                    height: {hSize};
                    border-width: 0px;
                    cursor: pointer;
                    background-image: url('{source}');
                    border-radius: 20px;
                    background-repeat: no-repeat;
                    background-size: cover;
                    background-position: center;
                    background-color: transparent;
                    transition: scale 0.2s ease-in-out;
                    {box_shadow}
                }}
                #{callback_id}:hover {{
                    scale: 0.9;
                }}
            </style>
            <div>
                <button id="{callback_id}"></button>
            </div>
            <script defer>
                document.querySelector("#{callback_id}").onclick = (e) => {{
                    google.colab.kernel.invokeFunction('{callback_id}', [], {{}})
                    e.preventDefault();
                }};
            </script>
        """
        html = template.format(
            title=self._title,
            callback_id=callback_id,
            source=self._source,
            wSize=self.wSize,
            hSize=self.hSize,
            tPos=self.tPos,
            lPos=self.lPos,
            box_shadow=box_shadow
        )
        return html



class InvokeBackground(object):
    def __init__(self, title, callback, source):
        self._title = title
        self._callback = callback
        self._source = source

    def _repr_html_(self):
        callback_id = 'background-' + str(uuid.uuid4())
        output.register_callback(callback_id, self._callback)

        template = """
            <style>
            body {{
                background-image: url('{source}');
                background-repeat: no-repeat;
                background-size: cover;
                background-position: center;
                height: 900px;
                width: 1200px;
            }}

            .background{{
                display: flex;
                justify-content: center;

            }}

            .headers{{
                font-size: 16px;
            }}
            </style>
            <div class='background'>
                <h1 class='headers>{title}</h1>
            </div>
        """
        html = template.format(title=self._title, callback_id=callback_id, source=self._source)
        return html

class InvokeBackgroundVideo(object):
    def __init__(self, title, callback, source):
        self._title = title
        self._callback = callback
        self._source = source

    def _repr_html_(self):
        callback_id = 'background-' + str(uuid.uuid4())
        output.register_callback(callback_id, self._callback)

        template = """
            <style>
            body {{
                background-color: black;
                height: 100vh;
                display: flex;
                justify-content: center;
                align-items: center;
            }}

            .background-video {{
                position: fixed;
                top: 0;
                left: 0;
                width: 100%;
                height: 100%;
                object-fit: cover;
            }}

            .headers {{
                font-size: 16px;
                color: white;
                text-align: center;
                z-index: 1;
                position: absolute;
                top: 50%;
                left: 50%;
                transform: translate(-50%, -50%);
            }}
            </style>
            <div class='background'>
                <video class='background-video' autoplay>
                    <source src='{source}' type='video/mp4'>
                </video>
                <h1 class='headers'>{title}</h1>
            </div>
        """
        html = template.format(title=self._title, callback_id=callback_id, source=self._source)
        # display(HTML(html))  ##  <-------    LIOR: fixed this  there was a retutn statement missing so i added it and it removed the black screen with the error
        return html


class InvokeInfoBackground(object):
    def __init__(self, title, callback, source):
        self._title = title
        self._callback = callback
        self._source = source

    def _repr_html_(self):
        callback_id = 'background-' + str(uuid.uuid4())
        output.register_callback(callback_id, self._callback)

        template = """
            <style>
            body {{
                background-image: url('{source}');
                background-repeat: no-repeat;
                background-size: auto;
                background-position: center;
                height: 900px;
                width: 1200px;
            }}

            .background{{
                display: flex;
                justify-content: center;

            }}

            .headers{{
                font-size: 16px;
            }}
            </style>
            <div class='background'>
                <h1 class='headers>{title}</h1>
            </div>
        """
        html = template.format(title=self._title, callback_id=callback_id, source=self._source)
        return html


class InvokeLogo(object):
    def __init__(self, title, callback, source):
        self._title = title
        self._callback = callback
        self._source = source

    def _repr_html_(self):
        callback_id = 'menu-' + str(uuid.uuid4())
        output.register_callback(callback_id, self._callback)

        template = """
            <style>
            .image-container{{
                width: 130px;
                height: 140px;
            }}
            </style>
            <image class="image-container" src={source}>
            </image>
        """
        html = template.format(title=self._title, callback_id=callback_id, source=self._source)
        return html


class InvokeLoginWindow(object):
    def __init__(self, title, callback, source):
        self._title = title
        self._callback = callback
        self._source = source

    def _repr_html_(self):
        callback_id_login = 'login-' + str(uuid.uuid4())
        output.register_callback(callback_id_login, self._callback)

        template = """
            <style>
              .loginmenu{{
                margin-top: 20px;
                width: 300px;
              }}

              .button-style{{
                position: absolute;
                top: 414px;
                left: 130px;
                width: 50px;
                height: 32px;
                background-color: transparent;
                border-width: 0px;
                transition: box-shadow 0.2s ease-in-out
              }}
              .button-style:hover{{
                box-shadow: 0 0 50px 6px rgba(54, 250, 0, 0.6);
              }}

              #username{{
                position: absolute;
                top: 295px;
                left: 48px;
                width: 212px;
                height: 25px;
                background-color: transparent;
                color: white;
                border-width: 0px;
                outline: none;
                font-size: 18px;
              }}

              #password{{
                position: absolute;
                top: 370px;
                left: 48px;
                width: 212px;
                height: 25px;
                background-color: transparent;
                color: white;
                border-width: 0px;
                outline: none;
                font-size: 18px;
              }}
            </style>

            <div>
            <image class="loginmenu" src={source}>
            </image>
            <div>
                <input type="text" placeholder="Enter Username" id="username">
                <input type="password" placeholder="Enter Password" id="password">
            </div>
            <button class="button-style" id="{callback_id_login}"></button>
            <br></br>
            </div>

            <script defer>
                document.querySelector("#{callback_id_login}").onclick = (e) => {{
                    var username = document.querySelector("#username").value;
                    var password = document.querySelector("#password").value;

                    if (username === "")
                        return

                    if (password === "")
                        return
                    google.colab.kernel.invokeFunction('{callback_id_login}', [username, password], {{}})

                }};
            </script>
        """

        html = template.format(title=self._title, callback_id_login=callback_id_login, source=self._source)
        return html


class InvokeMapButtons(object):
    def __init__(self, title, callback):
        self._title = title
        self._callback = callback

    def _repr_html_(self):
        callback_id = 'button-' + str(uuid.uuid4())
        output.register_callback(callback_id, self._callback)

        template = """
            <style>
                .{callback_id} {{
                    position: absolute;
                    border-width: 0px;
                    cursor: pointer;
                    background-color: transparent;
                    border-radius: 50px;
                }}

                #devops{{
                    background-repeat: no-repeat;
                    background-size: contain;
                    background-position: center;
                    top: -49px;
                    left: 221px;
                    width: 658px;
                    height: 658px;
                }}
                #devops:hover{{
                    background-image: url('https://github.com/audiblemaple/collab_test/blob/main/devops_btn.png?raw=true');
                }}

                #python{{
                    background-repeat: no-repeat;
                    background-size: contain;
                    background-position: center;
                    top: -36px;
                    left: 1006px;
                    width: 555px;
                    height: 555px;
                }}
                #python:hover{{
                    background-image: url('https://github.com/audiblemaple/collab_test/blob/main/python_btn.png?raw=true');
                }}

                #json{{
                    background-repeat: no-repeat;
                    background-size: contain;
                    background-position: center;
                    top: 400px;
                    left: 240px;
                    width: 590px;
                    height: 590px;
                }}
                #json:hover{{
                    background-image: url('https://github.com/audiblemaple/collab_test/blob/main/json_btn.png?raw=true');
                }}

                #xml{{
                    background-repeat: no-repeat;
                    background-size: contain;
                    background-position: center;
                    top: 417px;
                    left: 813px;
                    width: 990px;
                    height: 645px;
                }}
                #xml:hover{{
                    background-image: url('https://github.com/audiblemaple/collab_test/blob/main/xml_btn.png?raw=true');
                }}
            </style>

            <button class="{callback_id}"  id="devops"></button>
            <button class="{callback_id}"  id="python"></button>
            <button class="{callback_id}"  id="json"></button>
            <button class="{callback_id}"  id="xml"></button>

            <script defer>
                var buttons = document.querySelectorAll('.{callback_id}');
                buttons.forEach(function(button) {{
                    button.addEventListener('click', function() {{
                        var id = button.id;
                        google.colab.kernel.invokeFunction('{callback_id}', [id], {{}});
                    }});
                }});
            </script>
        """
        html = template.format(title=self._title, callback_id=callback_id)
        return html




class InvokeQuestion(object):
    def __init__(self, question, option1, option2, option3, option4, correct, question_type, callback, source, img_source):
        self._question      = question
        self._callback      = callback
        self._source        = source
        self._option1       = option1
        self._option2       = option2
        self._option3       = option3
        self._option4       = option4
        self._correct       = correct
        self._question_type = question_type
        self._img_source    = img_source
        self._image_uid     = 'img-' + str(uuid.uuid4())

    def _repr_html_(self):
        callback_id = 'menu-' + str(uuid.uuid4())
        output.register_callback(callback_id, self._callback)

        template = """
            <style>
                #mainDiv{{
                    background-image: url('{source}');
                    background-repeat: no-repeat;
                    background-size: contain;
                    background-position: center;
                    position: absolute;
                    width: 700px;
                    height: 450px;
                    top:40%;
                    align-items: flex-start;
                    display: flex;
                    flex-direction: column;
                    gap: 30px;
                    color: white;
                    margin-left: 20px;
                    padding-top: 100px;
                    padding-left: 100px;
                }}
                .{callback_id}{{
                    background-color: rgba(17, 84, 0, 0.5);
                    border-radius: 5px;
                    height: 30px;
                    color: white;
                    transition: scale 0.3s ease-in-out;
                }}
                .{callback_id}:hover{{
                    scale: 1.09;
                }}

            </style>
            <image id="{image_uid}" src="{img_source}" style="height:300px; width:570px; padding-top: 20px; padding-left: 40px;" onerror="this.style.display='none'"></image>
            <div id=mainDiv>
                <h2 style="color: white; white-space: normal; width: 620px;">{question}</h2>
                <button class="{callback_id}" >{option1}</button>
                <button class="{callback_id}" >{option2}</button>
                <button class="{callback_id}" >{option3}</button>
                <button class="{callback_id}" >{option4}</button>
            </div>

            <script defer>
                var buttons = document.querySelectorAll('.{callback_id}');
                buttons.forEach(function(button) {{
                    button.addEventListener('click', function() {{
                        var answer = button.innerText;
                        document.getElementById('{image_uid}').style.display = 'none';
                        google.colab.kernel.invokeFunction('{callback_id}', [answer, '{correct}', '{question_type}'], {{}});
                    }});
                }});
            </script>
        """
        html = template.format(question=self._question, option1=self._option1, option2=self._option2, option3=self._option3, option4=self._option4, correct=self._correct, question_type=self._question_type, callback_id=callback_id, source=self._source, img_source=self._img_source, image_uid=self._image_uid)
        return html






## DEPRECATED!!!!
# class InvokeLives(object):
#     def __init__(self, lives):
#         self.lives = lives

#     def _repr_html_(self):
#         template = """
#             <style>
#                 .lives{{
#                     position: absolute;
#                     top: 270px;
#                     right: 55px;
#                     color: red;
#                 }}
#             </style>

#             <h2 class="lives"></h2>

#             <script defer>
#                 var elements = document.getElementsByClassName('lives');
#                 for (var i = 0; i < elements.length; i++) {{
#                     elements[i].innerText = "Lives: {lives}";
#                 }}
#             </script>
#         """
#         html = template.format(lives=self.lives)
#         return html



# add positional argumenst, also change style... !!!!!!!
# add timeout control


class InvokePopUp(object):
    active_popup = None  # Track the active popup instance

    def __init__(self, text, color):
        self.text = text
        self.color = color
        self.messageuuid = 'message-' + str(uuid.uuid4())

    def _repr_html_(self):
        if self.color == 'red':
            background_style = 'background-image: url("https://github.com/audiblemaple/collab_test/blob/main/rederror.png?raw=true"); background-size: cover;'
        elif self.color == 'green':
            background_style = 'background-image: url("https://github.com/audiblemaple/collab_test/blob/main/greenerror.png?raw=true"); background-size: cover;'

        template = """
            <style>
                .centered {{
                    position: fixed;
                    top: 50%;
                    left: 50%;
                    transform: translate(-50%, -50%);
                    {background_style}
                    padding: 20px;
                    width: 400px; /* Adjust the container width */
                    height: 200px; /* Adjust the container height */
                    text-align: center;
                    display: flex;
                    align-items: center;
                    justify-content: center;
                    color: white;
                    opacity: 0;
                    animation: fadeInOut 1.5s forwards, growShrink 1.5s forwards;
                }}

                @keyframes fadeInOut {{
                    0% {{ opacity: 0; }}
                    25% {{ opacity: 1; }}
                    75% {{ opacity: 1; }}
                    100% {{ opacity: 0; }}
                }}

                @keyframes growShrink {{
                    0% {{ transform: translate(-50%, -50%) scale(0); }}
                    25% {{ transform: translate(-50%, -50%) scale(1); }}
                    75% {{ transform: translate(-50%, -50%) scale(1); }}
                    100% {{ transform: translate(-50%, -50%) scale(0); }}
                }}

                .centered-text {{
                    font-size: 24px; /* Adjust the font size */
                    font-family: cursive; /* Specify the cursive font family */
                    text-shadow: 2px 2px 4px rgba(0, 0, 0, 0.5); /* Add the text shadow effect */
                }}
            </style>

            <div id="{messageuuid}" class="centered">
                <p class="centered-text">{text}</p>
            </div>

            <script defer>
                // Get the reference to the div element
                var myDiv = document.getElementById("{messageuuid}");

                setTimeout(function() {{
                    myDiv.classList.add("fade-in");
                }}, 100);

                setTimeout(function() {{
                    myDiv.style.display = "none";
                }}, 2000);
            </script>
        """
        html = template.format(
            text=self.text,
            background_style=background_style,
            color=self.color,
            messageuuid=self.messageuuid
        )
        return html

    def show(self):
        if InvokePopUp.active_popup is not None:
            # If there's an active popup, wait for its fade-out animation to complete
            return

        InvokePopUp.active_popup = self  # Set the current instance as active
        display(HTML(self._repr_html_()))

        def hide_popup():
            self._hide()
            InvokePopUp.active_popup = None  # Reset the active popup

        # Hide the popup after 2 seconds
        threading.Timer(2, hide_popup).start()

    def _hide(self):
        # Code to hide the popup
        javascript = f'document.getElementById("{self.messageuuid}").style.display = "none";'
        display(Javascript(javascript))







def show_login_screen():
    output.clear()
    global question_number
    question_number = 0
    background = InvokeBackground('background', None, 'https://github.com/audiblemaple/collab_test/blob/main/main.png?raw=true')
    logo = InvokeLogo('background', None, 'https://github.com/audiblemaple/collab_test/blob/main/logo.png?raw=true')
    login = InvokeLoginWindow('background', log_user_in, 'https://github.com/audiblemaple/collab_test/blob/main/login.png?raw=true')
    exit = InvokeButtonWithPlace('', exit_app, 'https://github.com/audiblemaple/collab_test/blob/main/X.png?raw=true', '100px', '100px', '3%', '92%')

    display(background, logo, login, exit)


def log_user_in(username, password):
    global user_key
    user_key = 'Null'
    users_ref = FIREBASE_CONNECTION.get('/users/', None)
    key = None
    for key in users_ref.keys():
        if users_ref[key]['username'] == username:
          if users_ref[key]['password'] == password:
            user_key = key
            if users_ref[key]['isManager'] == '1':
                show_main_container()
                break
            show_user_screen()
            break
    if user_key == 'Null':
        message = InvokePopUp("user does not exist", "red")
        message.show()
        #display(message)



def show_user_screen():
    output.clear()
    background = InvokeBackground('background', None, 'https://github.com/audiblemaple/collab_test/blob/main/main.png?raw=true')
    logo = InvokeLogo('', None, 'https://github.com/audiblemaple/collab_test/blob/main/logo.png?raw=true')
    play = InvokeButtonWithPlace('', show_map_screen, 'https://github.com/audiblemaple/collab_test/blob/main/play.png?raw=true', '300px', '80px', '20%', '3%')
    my_progress = InvokeButtonWithPlace('', show_leaderboards, 'https://github.com/audiblemaple/collab_test/blob/main/scores.png?raw=true', '300px', '80px', '32%', '3%')
    my_scores = InvokeButtonWithPlace('', score_graph, 'https://github.com/audiblemaple/collab_test/blob/main/progress.png?raw=true', '300px', '80px', '44%', '3%')


    exit = InvokeButtonWithPlace('', exit_app, 'https://github.com/audiblemaple/collab_test/blob/main/X.png?raw=true', '100px', '100px', '3%', '92%')
    info = InvokeButtonWithPlace('', info_screen, 'https://github.com/audiblemaple/collab_test/blob/main/info.png?raw=true', '100px', '100px', '3%', '84%')

    # here we need to get the players data and show him the best and last score
    display(background, logo, my_progress, my_scores, play, exit, info)


def show_leaderboards():
    html = '''
      <style>
        #leaderboard{
            background-image: url("https://github.com/audiblemaple/collab_test/blob/main/scroll.png?raw=true");
            background-repeat: no-repeat;
            background-size: cover;
            background-position: center;
            position: absolute;
            top: 170px;
            left: 380px;
            width: 700px;
            height: 400px;
            padding: 50px;
            border-radius: 80px;
        }


        table {
            width: 100%;
            border-collapse: collapse;
            margin: 25px 0;
            font-size: 16px;
            font-family: comic Sans ms;
            font-weight: bold;
            color: black;
        }

        th, td {
          border: none;
          padding: 8px;
          text-align: center;
          border-bottom: 1px solid black;
        }

        th {
            border: 1px solid black
        }

      </style>

      <div id="leaderboard">
        <table>
          <tr>
            <th>Username</th>
            <th>Python</th>
            <th>XML</th>
            <th>JSON</th>
            <th>DevOps</th>
            <th>Total Score</th>
          </tr>
    '''

    users_ref = FIREBASE_CONNECTION.get('/users/', None)
    # sort the users by total high score descending
    sorted_users = sorted(users_ref.values(), key=lambda x: int(x['sumHighScore']), reverse=True)
    # Populate the table with data
    for user in sorted_users:
        if user['isManager'] == '1':
            continue
        html += f'''
        <tr>
            <td>{user['username']}</td>
            <td>{user['pythonHighScore']}</td>
            <td>{user['xmlHighScore']}</td>
            <td>{user['jsonHighScore']}</td>
            <td>{user['devopsHighScore']}</td>
            <td>{user['sumHighScore']}</td>
        </tr>
        '''

    html += '''
      </table>
    </div>
    '''
    exit = InvokeButtonWithPlace('', show_user_screen, 'https://github.com/audiblemaple/collab_test/blob/main/X.png?raw=true', '50px', '50px', "165px", "1100px")

    display(HTML(html), exit)

def show_map_screen():
    output.clear()
    background = InvokeBackground('background', None, 'https://github.com/audiblemaple/collab_test/blob/main/map.png?raw=true')
    quiz_buttons = InvokeMapButtons('', start_quiz)
    back_button = InvokeButtonWithPlace('', show_user_screen, 'https://github.com/audiblemaple/collab_test/blob/main/home.png?raw=true', '100px', '100px', '3%', '92%')
    info = InvokeButtonWithPlace('rocks', info_screen_fromMap, '', '100px', '100px', '35%', '48%')
    #info = InvokeButtonWithPlace('rocks', info_screen_fromMap, 'https://github.com/audiblemaple/collab_test/blob/main/map_info.png?raw=true', '100px', '100px', '35%', '48%')

    credits = InvokeButtonWithPlace('', credits_fire, 'https://github.com/audiblemaple/collab_test/blob/main/credits.png?raw=true', '100px', '100px', '3%', '85%')
    display(background, quiz_buttons, back_button , info, credits)






def show_quiz_video(background_video):
    output.clear()
    background = InvokeBackgroundVideo('', None, background_video)
    display(background)


def score_graph():
    global user_key
    user_name = FIREBASE_CONNECTION.get('/users', user_key)['username']

    session_data = FIREBASE_CONNECTION.get(f'/users/{user_key}/sessions/',None);
    if session_data == None:
        return None
    sums = {
            'devops':0,
            'python':0,
            'json':0,
            'xml':0
            }
    for key in session_data.keys():
        topic = session_data[key]['topic']
        score = session_data[key]['score']
        sums[topic] += score;

    x_topics = ['DevOps', 'Python', 'JSON', 'XML']
    y_scores = [sums['devops'],sums['python'],sums['json'],sums['xml']]

    fig = go.Figure()
    fig.add_trace(go.Bar(x=x_topics,y=y_scores))
    fig.update_layout(
        title=f'Sessions summary for {user_name}',
        xaxis_title="Topic",
        yaxis_title="Total Score",
        showlegend=False,
        legend_title="Legend",
        barmode='stack',
        bargap=0.2,
        bargroupgap=0.1,
        template='plotly_dark'
    )
    fig.update_xaxes(type='category')
    fig.show()

    exit = InvokeButtonWithPlace('', show_user_screen, 'https://github.com/audiblemaple/collab_test/blob/main/X.png?raw=true', '100px', '100px', '10%', '63%')

    background = InvokeBackground('background', None, 'https://github.com/audiblemaple/collab_test/blob/main/main.png?raw=true')
    display(background, exit)



def graph_builder(details):
    subject_name = quiz_topic
    true = "\ntrue"
    false = "\nfalse"
    y_questions = [ details[1][1][0],details[1][1][1], details[2][1][0],details[2][1][1], details[3][1][0],details[3][1][1], details[4][1][0],details[4][1][1]]
    labels = [details[1][0]+ true ,details[1][0] + false ,details[2][0] + true ,details[2][0] + false, details[3][0] + true, details[3][0] + false, details[4][0] + true, details[4][0] + false]

    fig = go.Figure()
    fig.update_yaxes(range=[0, 5])  # Set the maximum value of the y-axis
    fig.add_trace(go.Bar(x=labels, y=y_questions, marker_color=['green', 'red', 'green', 'red', 'green', 'red', 'green', 'red']))
    fig.update_layout(
        title="Results for " + subject_name + " game:",
        xaxis_title="Question",
        yaxis_title="Question count",
        showlegend=False,
        legend_title="",
        barmode='stack',
        bargap=0.2,
        bargroupgap=0.1,
        template='plotly_dark'
    )
    fig.update_xaxes(type='category')
    fig.show()

    background = InvokeBackground('background', None, 'https://github.com/audiblemaple/collab_test/blob/main/main.png?raw=true')
    display(background)


# Update show_result_screen() function
def show_result_screen():
    output.clear()
    global result_matrix
    items = [""]
    i = 0
    for key, value in result_matrix.items():
        items.append([key, value])
    graph_builder(items)
    upload_score()
    sum_scores()
    quiz_topic = "Null"
    exit = InvokeButtonWithPlace('', show_map_screen, 'https://github.com/audiblemaple/collab_test/blob/main/X.png?raw=true', '100px', '100px', '0%', '63%')
    display(exit)


def upload_score():
    global quiz_topic
    global user_key
    global current_score

    user_data = FIREBASE_CONNECTION.get('/users', user_key)
    # add the session to the user sessions object
    session = {
        'date': date.today(),
        'score': current_score,
        'topic': quiz_topic,
    }
    FIREBASE_CONNECTION.post(f'/users/{user_key}/sessions/', session)

    # update high score
    high_score_field = {
        "devops": "devopsHighScore",
        "python": "pythonHighScore",
        "json": "jsonHighScore",
        "xml": "xmlHighScore",
    }
    if quiz_topic in high_score_field:
        field_name = high_score_field[quiz_topic]
        user_ref = {
            field_name: current_score,
        }
        if current_score > user_data[field_name]:
          FIREBASE_CONNECTION.patch(f'/users/{user_key}', user_ref)
    else:
      print("CANT UPLOAD TO DATABASE!!!")


def sum_scores():
    user_data = FIREBASE_CONNECTION.get('/users', user_key)
    highscore = user_data['devopsHighScore'] + user_data['pythonHighScore'] + user_data['jsonHighScore'] + user_data['xmlHighScore']
    user_ref = {
        'sumHighScore': highscore,
    }
    FIREBASE_CONNECTION.patch(f'/users/{user_key}', user_ref)
    return




def query_quiz(question_list, quiz_type):
    for question, ans1, ans2, ans3, ans4, correct, qtype, pic in question_generator(question_list):
        pic_src = None
        if pic != "null":
            pic_data = base64.b64decode(pic)
            pic_src = f"data:image/png;base64,{base64.b64encode(pic_data).decode()}"
        match quiz_type:
            case "devops":
                question = InvokeQuestion(question, ans1, ans2, ans3, ans4, correct, qtype, eval_answer, "https://github.com/audiblemaple/collab_test/blob/main/qboxdevops.png?raw=true", pic_src)
            case "python":
                question = InvokeQuestion(question, ans1, ans2, ans3, ans4, correct, qtype, eval_answer, "https://github.com/audiblemaple/collab_test/blob/main/qboxpython.png?raw=true", pic_src)
            case "json":
                question = InvokeQuestion(question, ans1, ans2, ans3, ans4, correct, qtype, eval_answer, "https://github.com/audiblemaple/collab_test/blob/main/qboxjson.png?raw=true", pic_src)
            case "xml":
                question = InvokeQuestion(question, ans1, ans2, ans3, ans4, correct, qtype, eval_answer, "https://github.com/audiblemaple/collab_test/blob/main/qboxxml.png?raw=true", pic_src)

        display(question)
        yield


def question_generator(question_list):
    keys = list(question_list.keys())
    random_keys = random.sample(keys, len(keys))
    for key in random_keys:
        correct = question_list[key]["Correct"]
        ans1 = question_list[key]["Option1"]
        ans2 = question_list[key]["Option2"]
        ans3 = question_list[key]["Option3"]
        ans4 = question_list[key]["Option4"]
        pic = question_list[key]["Pic"]
        qtype = question_list[key]["Type"]
        question = question_list[key]["Question"]
        yield (question, ans1, ans2, ans3, ans4, correct, qtype, pic)



## check if i call next twice and reduce the number of question by 2 (VERY sure im not but need to check) ##################################################################
def fetch_questions(quiz_type):
    global generator
    global question_number
    question_number = 0
    quiz_to_fetch = '/pyAmericanQuestions/' + quiz_type +'/'
    question_list = FIREBASE_CONNECTION.get(quiz_to_fetch, None)
    generator = query_quiz(question_list, quiz_type)
    next(generator) #########

score_label = None

def show_quiz_screen(background, audio_file):
    output.clear()
    global current_score
    global score_label
    global question_number
    global result_matrix
    result_matrix = {}
    question_number = 0
    current_score = 0
    background = InvokeBackground('background', None, background)
    audio_html = """
        <audio id="audio" loop autoplay>
            <source src="{audio_source}" type="audio/mp3">
        </audio>
    """
    audio_html = audio_html.format(audio_source=audio_file)
    score_label = InvokeScore(str(current_score), question_number)
    back = InvokeButtonWithPlace('', show_map_screen, 'https://github.com/audiblemaple/collab_test/blob/main/goback.png?raw=true', '100px', '100px', '3%', '92%')
    display(background, back)
    audio_banner_html = """
        <div style="display: flex; align-items: center;">
            <button class="audio-control mute-btn" onclick="toggleMute(this)">Mute</button>
            <input class="audio-control volume-slider" type="range" min="0" max="100" value="75" onchange="adjustVolume(this.value)">
        </div>
    """
    display(HTML(audio_html))
    display(HTML("""
        <style>
            .audio-control {
                background-color: white;
                color: black;
                border: none;
                padding: 5px 10px;
                margin-right: 10px;
                font-weight: bold;
                cursor: pointer;
            }
            .volume-slider {
                width: 100px;
                height: 10px;
            }
        </style>
    """))
    display(HTML(audio_banner_html))

    # Call display_progress after the necessary elements are displayed
    score_label.display_progress()

    display(HTML("""
        <script>
            var audioElement = document.getElementById('audio');
            audioElement.volume = 0.50;
            audioElement.muted = false;

            function toggleMute(element) {
                audioElement.muted = !audioElement.muted;
                if (audioElement.muted) {
                    element.textContent = 'Unmute';
                } else {
                    element.textContent = 'Mute';
                }
            }

            function adjustVolume(value) {
                var volume = value / 100;
                audioElement.volume = volume;
            }
        </script>
    """))


from IPython.display import update_display


class InvokeScore(object):
    def __init__(self, score, progress_value):
        self.score = score
        self.progress_value = progress_value
        self.progress_statuses = []  # New field to keep track of correct and wrong answers
        self.progress_max = 20
        self.score_element = None
        self.progress_element = None

    def display_progress(self):
        self.score_element = display(HTML(self._create_score_html()), display_id=True)
        self.progress_element = display(HTML(self._create_progress_html()), display_id=True)

    def update_score_progress(self, score, progress_value, status=None):
        self.score = score
        self.progress_value = progress_value
        if status is not None:
            self.progress_statuses.append(status)  # update status
        if self.score_element is not None:
            self.score_element.update(HTML(self._create_score_html()))
        if self.progress_element is not None:
            self.progress_element.update(HTML(self._create_progress_html()))

    def _create_score_html(self):
        html = """
        <style>
            @import url('https://assets.codepen.io/9632/RocherColorGX.woff2');
            .score {{
                position: fixed;
                bottom: 40%;
                left: 65%;
                font-family: 'Rocher', cursive;
                font-size: 40px;
                display: flex;
                color: white;
                flex-direction: column;
                align-items: center;
                justify-content: center;
                margin: 0;
            }}
        </style>

        <div class="score">
            Score: {score}
        </div>
        """.format(score=self.score)
        return html

    def _create_progress_html(self):
        segment_width = 100 / self.progress_max  # Each segment will occupy this percentage of total width
        html = """
        <style>
            .progress-container {{
                position: fixed;
                bottom: 30%;
                left: 70%;
                transform: translateX(-50%);
                width: 600px;
                height: 30px;
                background-color: rgba(0, 0, 0, 0.5);
                border-radius: 5px;
                border: 1px solid white;
                box-shadow: 0 2px 5px rgba(0, 0, 0, 0.25);
                padding: 5px;
                display: flex;
            }}

            .progress-bar {{
                width: {segment_width}%;  /* Each segment's width is set to segment_width */
                height: 30px;
                border-radius: 5px;
                transition: width 0.5s ease-in-out;
            }}

            .correct {{
                background-color: #00b300;
            }}

            .wrong {{
                background-color: #b30000;
            }}
        </style>

        <div class="progress-container">
        """.format(segment_width=segment_width)

        for status in self.progress_statuses:
            html += f'<div class="progress-bar {status}"></div>'

        html += '</div>'

        return html



def eval_answer(user_answer, correct_answer, question_type):
    global generator
    global current_score
    global question_number
    global result_matrix
    global score_label

    if question_type not in result_matrix:
        result_matrix[question_type] = [0, 0]

    if user_answer == correct_answer:
        result_matrix[question_type][0] += 1
        current_score += 10
        question_number += 1
        score_label.update_score_progress(str(current_score), question_number, status='correct')  # status added
        message = InvokePopUp("Good job!", "green")
        message.show()
        time.sleep(1.5)
    else:
        message = InvokePopUp("You're a noob LOL", "red")
        result_matrix[question_type][1] += 1
        question_number += 1
        score_label.update_score_progress(str(current_score), question_number, status='wrong')  # status added
        message.show()
        time.sleep(1.5)

    try:
        next(generator)
    except StopIteration:
        show_result_screen()
        current_score = 0
        question_number = 0
    return



def start_quiz(quiz_name):
    global quiz_topic
    match quiz_name:
        case "devops":
            show_quiz_video('https://github.com/audiblemaple/collab_test/blob/main/p03.mp4?raw=true')
            time.sleep(12)
            show_quiz_screen("https://github.com/audiblemaple/collab_test/blob/main/devops.png?raw=true", "https://github.com/audiblemaple/collab_test/blob/main/p03.mp3?raw=true")
        case "python":
            show_quiz_video('https://github.com/audiblemaple/collab_test/blob/main/leshy.mp4?raw=true')
            time.sleep(12)
            show_quiz_screen("https://github.com/audiblemaple/collab_test/blob/main/py.png?raw=true",'https://github.com/audiblemaple/collab_test/blob/main/leshy.mp3?raw=true')
        case "json":
            show_quiz_video('https://github.com/audiblemaple/collab_test/blob/main/magnificus.mp4?raw=true')
            time.sleep(12)
            show_quiz_screen("https://github.com/audiblemaple/collab_test/blob/main/json.png?raw=true", 'https://github.com/audiblemaple/collab_test/blob/main/magnificus.mp3?raw=true')
        case "xml":
            show_quiz_video('https://github.com/audiblemaple/collab_test/blob/main/grimora.mp4?raw=true')
            time.sleep(12)
            show_quiz_screen("https://github.com/audiblemaple/collab_test/blob/main/xml.png?raw=true", 'https://github.com/audiblemaple/collab_test/blob/main/Grimora.mp3?raw=true')
    quiz_topic = quiz_name
    fetch_questions(quiz_name)


def click_test(param, param2):
    print(param, " = ", param2)



def info_screen():
    info_background = InvokeInfoBackground('background', None, 'https://github.com/audiblemaple/collab_test/blob/main/gameInfoScreen.png?raw=true')
    back = InvokeButtonWithPlace('', show_user_screen, 'https://github.com/audiblemaple/collab_test/blob/main/X.png?raw=true', '100px', '100px', '10%', '65%')
    display(info_background,back)

def info_screen_fromMap():
    output.clear()
    info_background = InvokeInfoBackground('background', None, 'https://github.com/audiblemaple/collab_test/blob/main/gameInfoScreen.png?raw=true')
    back = InvokeButtonWithPlace('', show_map_screen, 'https://github.com/audiblemaple/collab_test/blob/main/X.png?raw=true', '100px', '100px', '10%', '65%')
    display(info_background,back)

def not_supported_yet():
    print("not supported yet")

def credits_fire():
    output.clear()
    show_quiz_video('https://github.com/audiblemaple/collab_test/blob/main/creditsFINAL.mp4?raw=true')
    time.sleep(23)
    show_map_screen()


def exit_app():
    output.clear()
    print("App canceled...")

# this is the first screen
show_login_screen()

```
